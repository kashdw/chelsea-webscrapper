```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
website_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
data = requests.get(website_url)
```

```python
data.text
```

```python
bs = BeautifulSoup(data.text)
```

```python
league_table = bs.select("table.stats_table")[0]
```

```python
league_table
```

```python
links = league_table.find_all("a")
```

```python
links
```

```python
links = [l.get("href") for l in links]
```

```python
links
```

```python
links = [l for l in links if '/squads' in l]
```

```python
links
```

```python
full_team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
full_team_urls
```

# to get only chelsea data

```python
chelsea_url = full_team_urls[2]
```

```python
chelsea_url
```

```python
data = requests.get(chelsea_url)
```

```python
data.text
```

```python
matches = pd.read_html(data.text, match = "Scores & Fixtures")
```
